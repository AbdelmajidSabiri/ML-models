```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
teta1,teta0 = 6,2
pointsX = np.linspace(10,30,40)
pointsY = teta1*pointsX + teta0

pointsYm = np.zeros(40)

for i in range(40) :
    pointsYm[i] = pointsY[i] + random.randint(-20,20)
    
plt.scatter(pointsX,pointsYm)
plt.plot(pointsX,pointsY)
```
