```python
import numpy as np 
from sklearn.datasets import make_blobs
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import mode

```

```python
X, y = make_blobs(n_samples=100, centers = 3,n_features=2, random_state=0)
x = [0,3]
plt.scatter(X[y==0][:,0],X[y==0][:,1], label="C0")
plt.scatter(X[y==1][:,0],X[y==1][:,1], label="C1")
plt.scatter(X[y==2][:,0],X[y==2][:,1], label="C2")
plt.scatter(x[0],x[1])
plt.legend()
```

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
def kppv(x,X,y,k):
    distances = [euclidean_distance(x, x_train) for x_train in X]
    
    df = pd.DataFrame({
        'x1': X[:, 0],
        'x2': X[:, 1],
        'distance': distances,
        'y': y
    })
    df_sorted = df.sort_values('distance')
    sorted_numpy = df_sorted.to_numpy()[:k]

    return mode(sorted_numpy[:,3])
```

```python
print(kppv(x,X,y,3))
```
