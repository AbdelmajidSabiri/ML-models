```python
import numpy as np
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
S = F=["ML","DM","SFE","Py"]
S=np.array([[5,8,7,12,0],
  [8,7,10,15,1],
  [12,10,17,17,2],
  [17,14,18,9,2],
   [10,17,16,9,2],
  [12.5,17,16,9,2],
  [5,6,12,10,0],
  [1,8,12,15,0],
  [3,10,16,16,1],
  [16,8,13,8,2],
  [14,9,16,9,2]]).astype("float32")
```

```python
X = S[:,:-1]
Y = S[:,-1]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
plt.figure(figsize=(12, 8)) 
tree.plot_tree(
    clf, 
    filled=True,            
    rounded=True,           
    feature_names=F,         
    class_names=["Class 0", "Class 1", "Class 2"],
    fontsize=10,          
    impurity=False,   
    proportion=True  
)
plt.show()
```

```python
x = [10,17,16,9]
clf.predict([x])
```
