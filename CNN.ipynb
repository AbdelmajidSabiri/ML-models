```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
```

```python
img = image.imread('images/cnn_img.jpg')
print(img.dtype)
print(img.shape)
plt.imshow(img[:,:,:])
```

```python
K = np.random.random((3,3))

def conv(M,K):
    C = np.empty_like(M[1:-1,1:-1])
    m,n = C.shape
    k, l = K.shape
    
    for i in range(m):
        for j in range(n):
            C[i,j] = np.mean(M[i:i+3,j:j+3] * K)
        
    return C
```

```python
img1 = img[:,:,0].astype("float32")/255
c = conv(img1, K)
c.shape
```

```python
plt.imshow(c)
```

```python
def maxpooling(M, d):
    S = int(d/2)
    m,n = M.shape
    P = np.empty((int(m/d), int(n/d)))
    
    for i in range(0,m/d, 2) :
        for j in range(n/d, 2) :
            P[i,j] = np.max(M[i:i+d, j:j+d])
            
    return P
```

```python
P = maxpooling(img1,2)
plt.imshow(P, cmap="grey")
```
